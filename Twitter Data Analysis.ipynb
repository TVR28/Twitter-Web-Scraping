```python
df1[df1.columns[3:]]=df1[df1.columns[3:]].fillna(0)
df1.isnull().sum()
```

```python
df1
```

```python
#change datatypes of features and clean the values
df1['reTweets'] = df1['reTweets'].astype(str).apply(lambda x:x.replace(',',""))
df1['Likes'] = df1['Likes'].astype(str).apply(lambda x:x.replace(',',""))
df1['Replys'] = df1['Replys'].astype(str).apply(lambda x:x.split('.')[0])
df1['Tweets'] = df1['Tweets'].astype(str).apply(lambda x:x.replace(',',""))
df1["UserTags"]=df1["UserTags"].astype(str)
df1["Tweets"]=df1["Tweets"].astype(str)
df1['TimeStamps'] = pd.to_datetime(df1['TimeStamps'])
df1['reTweets'] = df1['reTweets'].apply(lambda x:x.replace(',',""))
df1[df1.columns[3:]]=df1[df1.columns[3:]].astype(int)
df1
```

```python
df1.duplicated().sum()
```

```python
df1.drop_duplicates(inplace=True,ignore_index=True)
df1.duplicated().sum()
```

```python
df1
```

```python
df1.rename(columns={'Replys' : 'Replies'},inplace=True)
df1
```

```python
df1["Account"] = df1["UserTags"].apply(lambda x: x.split('\n')[0])
df1["Handle"] = df1["UserTags"].apply(lambda x: x.split('\n')[1])
df1
```

```python
df1.sort_values("TimeStamps")
df1
```

```python
df1["TimeStamps"]=df1["TimeStamps"].astype(str)
```

```python
df1['Date'] = df1['TimeStamps'].apply(lambda x: x.split(' ')[0])
df1['Time'] = df1['TimeStamps'].apply(lambda x: x.split(' ')[1])
df1['Time'] = df1['Time'].apply(lambda x: x.split('+')[0])
df1
```

```python
df1.drop(columns ={"UserTags","TimeStamps"},inplace=True)
df1
```

```python
df1.info()
```

```python
df1=df1.reindex(columns=['Account','Handle','Date','Time','Tweets','Replies','reTweets','Likes'])
df1
```
